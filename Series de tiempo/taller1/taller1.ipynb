```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Parámetros de la distribución exponencial
lam = 0.5 # tasa de decaimiento
n = 1000 # longitud de la serie temporal

# Generar serie temporal de ruido blanco discreto
s = np.random.exponential(scale=1/lam, size=n)

# Hacer un subplot de 1x3 con la serie temporal, el histograma y el ACF usando plot_acf
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(s)
ax[0].set_title('Serie temporal')
ax[1].hist(s, bins=20)
ax[1].set_title('Histograma')
plot_acf(s, ax=ax[2], lags=20)
ax[2].set_title('ACF')
plt.show()
```
